# Example of the environment with a priority function

## Self play example

```python
from src.envs import NoThreeCollinearEnvWithPriority, Point
import numpy as np
```

```python
env = NoThreeCollinearEnvWithPriority(3, 5)
env.self_play_add_point(Point(1,1))
env.self_play_add_point(Point(2,1))
env.self_play_add_point(Point(1,2))
```

## Greedy Search with a Priority Function

```python
def priority(p: Point, grid_size) -> float:
    x, y = p.x, p.y
    m,n = grid_size
    return max(x, y) / ((n - 1)*(m - 1))

env = NoThreeCollinearEnvWithPriority(5, 5, priority)

result = env.greedy_search()
print(f"found {result} points during the greedy search with priority function")
env.plot()
```
